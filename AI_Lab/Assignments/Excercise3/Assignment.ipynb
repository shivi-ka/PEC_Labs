4. Data Visualization-Create a histogram for each feature in the dataset to illustrate the feature distributions. Plot each histogram. 

```python
import matplotlib.pyplot as plt
plt.hist(df['SepalLengthCm'])
plt.title('Distribution of Sepal Length (cm)')
plt.show()
plt.hist(df['SepalWidthCm'])
plt.title('Distribution of Sepal Width (cm)')
plt.show()
plt.hist(df['PetalLengthCm'])
plt.title('Distribution of Petal Length (cm)')
plt.show()
plt.hist(df['PetalWidthCm'])
plt.title('Distribution of Petal Width (cm)')
plt.show()
```

5. Identify the relationship between sepal-length and sepal-width, also petal-length and petal-width. Show it using scatterplot. Place the legend outside the figure.

```python
plt.scatter(data=df, x='SepalLengthCm', y='SepalWidthCm')
plt.title('Relationship between Sepal Length & Width')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
plt.scatter(data=df, x='PetalLengthCm', y='PetalWidthCm')
plt.title('Relationship between Petal Length & Width')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

6. Create a boxplot for each feature in the dataset. All of the boxplots should be combined into a single plot. Compare distributions.

```python
data_to_plot = [df['SepalLengthCm'], df['SepalWidthCm'], df['PetalLengthCm'], df['PetalWidthCm']]
plt.boxplot(data_to_plot)
plt.grid()
plt.show()
```

7. Identify outliers. Handle those outliers.

```python
import numpy as np

plt.boxplot(df['SepalWidthCm'])
plt.grid()
plt.show()
Q1 = np.percentile(df['SepalWidthCm'], 25, interpolation = 'midpoint')
Q3 = np.percentile(df['SepalWidthCm'], 75, interpolation = 'midpoint')
IQR = Q3 - Q1
# Upper bound
upper = np.where(df['SepalWidthCm'] >= (Q3+1.5*IQR))
  
# Lower bound
lower = np.where(df['SepalWidthCm'] <= (Q1-1.5*IQR))
  
# Removing the Outliers
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)
plt.boxplot(df['SepalWidthCm'])
plt.grid()
plt.show()
```
