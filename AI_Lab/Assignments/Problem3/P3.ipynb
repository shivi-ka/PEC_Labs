## Load the data file Titanic_original.csv

1. Create a pie chart presenting the male/female proportion

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("titanic_original.csv")
count = data['sex'].value_counts().to_dict()
plt.pie(count.values(), labels=count.keys(), autopct='%.2f%%')
plt.show()
```

2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("titanic_original.csv")
plt.scatter(data[data.sex == 'male'].age, data[data.sex == 'male'].fare, c='#0047ed')
plt.scatter(data[data.sex == 'female'].age, data[data.sex == 'female'].fare, c='#f70968')
plt.xlabel('age')
plt.ylabel('fare')
plt.legend(['male', 'female'])
plt.show()
```
