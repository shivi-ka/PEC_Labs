### Write a python script to display bar plot for the “Sales.csv” file with column Day on x axis,  sales (in Rs) on y axis for 3 weeks and having the following customization:
1. Color of bar blue(Week 1), red(Week 2), green(Week 3)
2. Keep edge color as purple
3. Line width as 2
4. Line style as “--”
5. Mention the legend

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('sales.csv')
df.plot(
    kind = 'bar',
    x = 'Day', 
    color = ['blue', 'red', 'green'], 
    edgecolor='purple', 
    linewidth=2, 
    linestyle='--'
)
plt.ylabel('Sales (in Rs.)')
plt.show()
```
