## Problem 2: Visualize the company sales (sales.csv) data using “matplotlib”. 
### a. Read Total profit of all months and show it using a line plot


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales.csv")
profitList = df['total_profit'].tolist()
monthList  = df ['month'].tolist()
plt.plot(
    monthList,
    profitList
)
plt.show()
```

### b. Get total profit of all months and show line plot with the following  properties
- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales.csv")
profitList = df['total_profit'].tolist()
monthList  = df ['month'].tolist()
plt.plot(
    monthList,
    profitList,
    linestyle='--',
    color='red',
    label = 'Profit Data for Year',
    marker = 'o',
    markerfacecolor = 'red',
    linewidth = 3
)
plt.legend(loc='lower right')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.show()
```

### c. Display the number of units sold per month for each product using multiline plots. 

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales.csv")
monthList  = df ['month'].tolist()
phones_data =  df ['phones'].tolist() 
laptops_data = df ['laptops'].tolist()
chargers_data = df ['chargers'].tolist()
cables_data = df ['cables'].tolist()
watches_data = df ['watches'].tolist()
headphones_data = df ['headphones'].tolist()
plt.plot(monthList, phones_data)
plt.plot(monthList, laptops_data)
plt.plot(monthList, chargers_data)
plt.plot(monthList, cables_data)
plt.plot(monthList, watches_data)
plt.plot(monthList, headphones_data)
plt.show()
```
